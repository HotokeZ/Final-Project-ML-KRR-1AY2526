# Anomaly Analysis Notebook

This notebook loads the pipeline outputs and provides interactive exploration of anomalous profiles, clusters, and example cases.

Run the cells to reproduce the visuals. If files are in a different path, update the paths accordingly.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

DATA_DIR = 'output'
scores_path = os.path.join(DATA_DIR, 'anomaly_scores.csv')
clusters_path = os.path.join(DATA_DIR, 'anomaly_clusters.csv')
labels_path = os.path.join(DATA_DIR, 'cluster_labels.md')

scores = pd.read_csv(scores_path)
clusters = pd.read_csv(clusters_path) if os.path.exists(clusters_path) else None
display(scores.head())

# Show counts of anomalies and clusters
print('Total rows:', len(scores))
print('Total anomalies:', scores['is_anomaly'].sum())
if clusters is not None:
    display(clusters['cluster'].value_counts().sort_index())

# Display human readable labels if present
if os.path.exists(labels_path):
    txt = open(labels_path, 'r', encoding='utf-8').read()
    display(Markdown(txt))

# For each cluster, plot the top categories for a few important fields
important = ['Gender','Country','Occupation','self_employed','family_history','treatment','mental_health_interview']
if clusters is not None:
    for c in sorted(clusters['cluster'].unique()):
        sub = clusters[clusters['cluster']==c]
        print(f'--- Cluster {c} (n={len(sub)})')
        for col in important:
            if col in sub.columns:
                vc = sub[col].value_counts(normalize=True).head(8)
                plt.figure(figsize=(6,3))
                sns.barplot(x=vc.values, y=vc.index, palette='muted')
                plt.xlabel('Proportion')
                plt.title(f'Cluster {c} — {col}')
                plt.tight_layout()
                display(plt.gcf())
                plt.close()

# Show example rows per cluster
if clusters is not None:
    for c in sorted(clusters['cluster'].unique()):
        print(f'### Examples from cluster {c}')
        display(clusters[clusters['cluster']==c].head(5))
```
